### Arithmetic Operators

|Operation         |Operator  |Example    |
|------------------|----------|-----------|
| Addition         | +        | `4 + 4`   |
| Subtraction      | -        | `9 - 1`   |
| Multiplication   | *        | `4 * 2`   |
| Division         | /        | `16 / 2`  |
| Exponentiation   | **       | `2 ** 3`  |


```python
3**2 # **0.5 is a square root
```

### Comparison Operators

| Operator | Description              | Example  |
|----------|--------------------------|----------|
| `==`     | Equal to                 | `1 == 1` |
| `!=`     | Not equal to             | `1 != 0` |
| `>`      | Greater than             | `1 > 0`  |
| `<`      | Less than                | `0 < 1`  |
| `>=`     | Greater than or equal to | `1 >= 1` |
| `<=`     | Less than or equal to    | `1 <= 1` |

```python
1<1
```

### Logical Operators

They allow us to combine or negate statements
- `OR` expand our possibilities
- `AND` limit our possibilities
- `NOT` flips the boolean value

```python
is_first_of_the_month = True
report_has_been_sent = False

should_process_report = is_first_of_the_month and not report_has_been_sent

print(should_process_report)
```

### String Methods

- `.lower` and `.upper`: convert the string to all lower or upper case
- `.strip`: remove any leading and trailing whitespace from the string
- `.isdigit`: test whether or not the string is a number
- `.split`: convert a string to a list
- `.join`: convert a list to a string

<div class="alert alert-info" role="alert">
 
<b>NOTE:</b> All of the string methods here do not modify the original string, they return a NEW string with the modification

</div>

```python
s = '   Hello, Codeup!   '
```

```python
s.lower()
```

```python
s.strip()
```

```python
s = 'hi my name is Kelsey'
new_list = s.split()
print (new_list)
```

```python
' - '.join(new_list)
```

## Dictionaries
Dictionaries are a way to group values together and give a name to each value. 
- The names for the values in a dictionary are called keys. 
- It is common to refer to a dictionary as a group of key-value pairs.

#### Dictionaries can be created by writing a *dictionary literal*
- Delimited by curly brackets `{ }`
- The keys are separated from the values with a colon `:` 
- Key-value pairs are separated by commas `,`

```python
my_school = {'name': 'codeup',
            'age': 10,
            'instructors':{'instructor_name': 'misty'}}
```

```python
my_school['age']
```

```python
my_school['instructors']['instructor_name']
```
